```python
# Imports
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pydataset 
from sklearn.cluster import KMeans
```

```python
# Carregar dataset a partir do pydata
iris = pydataset.data('iris')
```

```python
# Mostrar as primeiras linhas
iris.head()
```

```python
# Definindo os eixos

x = iris[['Sepal.Width', 'Sepal.Length', 'Petal.Width', 'Petal.Length']]
y = iris['Species'].apply(lambda x: {'setosa': 0, 'virginica': 1, 'versicolor':
2}[x])
```

```python
# Definindo cores
colormap = np.array(['red', 'lime', 'black'])
```

```python
# Plots

plt.figure(figsize=(18, 8))
plt.subplot(1, 2, 1)
plt.scatter(x['Petal.Length'], x['Petal.Width'], 
            c=colormap[y])
plt.title('Petal')
plt.subplot(1, 2, 2)
plt.scatter(x['Sepal.Length'], x['Sepal.Width'], 
            c=colormap[y])
plt.title('Sepal')
```

```python
# Aplicando o Kmeans para 3 clusters

kmeans = KMeans(n_clusters=3, n_init=10)
kmeans.fit(x)

kmeans.labels_
```

```python
# Plot 3 clusters
plt.figure(figsize=(18, 8))
plt.subplot(1, 2, 1)
plt.scatter(x['Petal.Length'], x['Petal.Width'], 
            c=colormap[kmeans.labels_])
plt.title('Petal')
plt.subplot(1, 2, 2)
plt.scatter(x['Sepal.Length'], x['Sepal.Width'], 
            c=colormap[kmeans.labels_])
plt.title('Sepal')
```

```python
# Aplicando o Kmeans para 2 clusters

kmeans = KMeans(n_clusters=2, n_init=10)
kmeans.fit(x)
```

```python
# Plot 2 clusters
plt.figure(figsize=(18, 8))
plt.subplot(1, 2, 1)
plt.scatter(x['Petal.Length'], x['Petal.Width'], 
            c=colormap[kmeans.labels_])
plt.title('Petal')
plt.subplot(1, 2, 2)
plt.scatter(x['Sepal.Length'], x['Sepal.Width'], 
            c=colormap[kmeans.labels_])
plt.title('Sepal')
```

```python
# Set padro de cores

colormap = np.array(['red', 'lime', 'black', 'white', 'blue'])
```
