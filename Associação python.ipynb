```python
#Imports
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
dataset = [
['leite', 'cebola', 'manteiga', 'feijão', 'ovos', 'iogurte'],
['sal', 'cebola', 'manteiga', 'feijão', 'ovos', 'iogurte'],
['leite', 'maça', 'feijão', 'ovos'],
['leite', 'arroz', 'pipoca', 'feijão', 'iogurte'],
['pipoca', 'cebola', 'feijão', 'sorvete', 'ovos']
]
```

```python
dataset
```

```python
# Transforma a ocorrencia do elemento em uma unica coluna sem repetição e retorna o booleano para a presença ou não do elemento em cada lista
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df
```

```python
# Usa o apriori
apriori(df, min_support=0.6, use_colnames=True)
freq_items = apriori(df, min_support=0.6, use_colnames=True)
freq_items['tamanho'] = freq_items['itemsets'].apply(lambda x: len(x))
freq_items
freq_items.columns = ['Suporte', 'ConjuntoItens', 'Tamanho']
freq_items
help(freq_items.apply)
help(apriori)
freq_items[freq_items['Suporte'] >= 0.6]
```

```python
freq_items.columns = ['support', 'itemsets', 'tamanho']
association_rules(freq_items, metric='confidence')
```
